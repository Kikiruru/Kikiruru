````python
df_iris['sepal length (cm)'].hist(bins=30)
plt.show()
````

````python
for class_no in iris.target.unique():
    df_iris['sepal length (cm)'].iloc[np.where(iris.target == class_no)[0]].hist(bins=30)
````

````python
for class_no in iris.target.unique():
    df_iris.loc[iris.target == class_no, 'sepal length (cm)'].hist(bins=30)
````

````python
for class_no in iris.target.unique():
    df_iris.loc[iris.target == class_no, 'sepal width (cm)'].hist(bins=30, figsize=(7,5))
````
