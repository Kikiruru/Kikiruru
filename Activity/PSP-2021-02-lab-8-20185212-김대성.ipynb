#### 예제 6. 다음 조건을 만족하는 영상을 화면에 표시하라.

1. 한 변의 길이가 21인 정사각형의 모든 색은 청녹색(cyan)이다.
2. 정사각형 내부에 중심의 위치가 (10, 10)이고 반지름이 5인 원의 내부의 모든 색은 노란색(yellow)이다.

참고: 중심이 $(x_c, y_c)$이고 반지름이 $r$인 원 내부는 $(x-x_c)^2 + (y-y_c)^2 \le r^2$으로 표현된다.

````python
img = np.full((21, 21, 3),(0, 255, 255), np.uint8) # 청녹색 정사각형
r = 5
xc = np.array([10, 10])

for y in range(21):
    for x in range(21):
        if (y - xc[0])**2 + (x - xc[1])**2 <= r**2:
            img[y, x] = (255, 255, 0)
plt.imshow(img)
plt.show()
#aliasing sampling이 충분하지 않을 때 나타나는 현상
#21개로 sampling
````

````python
img = np.full((121, 121, 3),(0, 255, 255), np.uint8)
r = 30
xc = np.array([60, 60])

for y in range(121):
    for x in range(121):
        if (y - xc[0])**2 + (x - xc[1])**2 <= r**2:
            img[y, x] = (255, 255, 0)
plt.imshow(img)
plt.show()
#121개로 sampling
````

#### 예제 11. image flip

````python
plt.imshow(face[:,::-1,:])
plt.show()
````

````python
plt.imshow(face[::-1,:,:])
plt.show()
````

````python
def mix_image(img1, img2, p):
    mixed = p * img1 + (1 - p)*img2
    mixed = np.clip(mixed, 0, 255) # clip : 255보다 크면 255로 0보다 작으면 0으로 바꿔줌
    return mixed.astype(np.uint8) # np.uint8로 데이터 타입을 바꿔줌
````

````python
plt.imshow(face)
plt.show()
````

````python
new_img = mix_image(face, face[:,::-1,:], 0.6)
plt.imshow(new_img)
plt.show()
````

#### 예제 12.
Numpy의 fancy indexing을 이용해서 2이상 100 이하의 소수 구하기

````python
from math import sqrt
prime_numbers =[]
N = 100
for n in range(2,N+1):
#   for i in range(2,n//2):
    for i in range(2,int(sqrt(n)+1)):
        if not n%i:
            break
    else:
        prime_numbers.append(n)
print(f"Found {len(prime_numbers)} prime numbers")
print(prime_numbers)
````

##### 위 프로그램을 numpy ndarray의 fancy indexing을 이용해서 다시 작성하라.

````python
N = 100
prime_numbers = np.arange(N+1)
mask = np.array([True]*(N+1))
mask[:2] = False
for i in range(2, N//2):
    mask[2*i::i] = False
prime_numbers[mask]
````

#### 예제 13. 
수 직선에 10 개의 점 A, B, C, D, E, F, G, H, I, J가 있다고 가정하자.

각 점의 좌표를 다음과 같이 list로 표현하였다.

```
points = [0, 52, 85, 113, 135, 154, 195, 223, 260, 340]
```
10개 점으로부터 각 점까지의 거리를 나타내는 2차원 배열 distances를 두 가지 방법으로 구하라.

1) for loop을 이용한 방법

2) Numpy 배열과 브로드캐스팅을 이용한 방법

````python
points = [0, 52, 85, 113, 135, 154, 195, 223, 260, 340]
points
````

````python
distance = []
temp = []
for n in range(len(points)):
    for k in points:
        temp.append((k - points[n]))
    distance.append(temp)
    temp = []

distance
````

````python
points = np.array(points)
distance = np.zeros(0, dtype = np.int8)
for n in range(len(points)):
    distance = np.append(distance, points - points[n])
distance = distance.reshape(len(points), len(points))
distance
````

#### 문제 1. 3원색 원 그리기
아래와 같은 그림을 그려라.

![RGB circles](rgb_circles.png "RGB circles")

````python
N = 300
img = np.full((N, N, 3),(0, 0, 0), np.uint8)
r = N//4
xr = np.array([N//3, N//2]) #red
xg = np.array([N//1.5, N//3]) # green
xb = np.array([N//1.5, N//1.5]) # blue

x = np.arange(0, N)
y = np.arange(0, N).reshape(N, 1)

inside = (y-xr[0])**2 + (x-xr[1])**2 <= r**2 # broadcasting
img[inside] = img[inside] + (255, 0, 0) 
inside = (y-xg[0])**2 + (x-xg[1])**2 <= r**2
img[inside] = img[inside] + (0, 255, 0) 
inside = (y-xb[0])**2 + (x-xb[1])**2 <= r**2
img[inside] = img[inside] + (0, 0, 255) 

plt.imshow(img)
plt.show()
````

#### 문제 2.

$[0, 1]$ 구간에 속한 난수 25개 만들어서 $5 \times 5$ 2차원 배열 x를 만들고, 각 행에서 0.5에 가장 가까운 수를 찾아서 1차원 배열을 2가지 방법으로 구하라.


1) for loop을 이용한 방법

2) Numpy 배열과 브로드캐스팅을 이용한 방법

다음 두 줄을 각 프로그램의 첫 줄에 적어라.

```
np.random.seed(1)
np.set_printoptions(precision=3, suppress=True)
```
x와 결과는 각각 다음과 같아야 한다.
```
[[0.417 0.72  0.    0.302 0.147]
 [0.092 0.186 0.346 0.397 0.539]
 [0.419 0.685 0.204 0.878 0.027]
 [0.67  0.417 0.559 0.14  0.198]
 [0.801 0.968 0.313 0.692 0.876]]
[0.417 0.539 0.419 0.559 0.313]
```

````python
np.random.seed(1)
np.set_printoptions(precision=3, suppress=True)

x = np.random.rand(5,5)
print(x)
````

````python
max_ = np.zeros(0, dtype = np.int8)
print(x)
for n in range(len(x)):
    temp = 1
    for k in x[n]:
        if(abs(k - 0.5)) < abs(temp - 0.5): #절대값 비교
            temp = k
        elif(k == 0.5):
            temp = k
            break
    max_ = np.append(max_, temp)
    
print(max_)
````

````python
np.random.seed(1)
np.set_printoptions(precision=3, suppress=True)

x = np.random.rand(5,5)
a = np.full((1, len(x)), 0.5) # (행 개수, 열 개수), 채울 값
print(x)

x1 = abs(x - a)
for n in range(len(x)):
    x1[n] = x1[n] <= np.min(x1[n])
x1 = x1 == 1

print(x[x1])
````
