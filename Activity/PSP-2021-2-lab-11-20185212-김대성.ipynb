##### 1.2

print(average) 명령을 실행하면 화면에 아래와 같이 표현된다.
```
         Iris-setosa  Iris-versicolor  Iris-virginica
꽃받침길이cm        5.006            5.936           6.588
꽃받침폭cm         3.418            2.770           2.974
꽃잎길이cm         1.464            4.260           5.552
꽃잎폭cm          0.244            1.326           2.026
```

````python
flower_names = []

for n in df['종']:
    if n not in flower_names:
        flower_names.append(n)

print(flower_names)
````

````python
average[flower_names[0]] = ss1
print(average)
````

````python
ss2 = df[df['종'] == flower_per_dept.index[1]].mean(axis=0)
print(ss2)
````

````python
ss3 = df[df['종'] == flower_per_dept.index[2]].mean(axis=0)
print(ss3)
````

### for문을 이용해 average 정리

````python
flower_names
````

````python
average = pd.DataFrame()

for n in range(3):
    average[flower_names[n]] = df[df['종'] == flower_names[n]].mean(axis=0)
    # df['종'] == flower_names인 것중에서 axis=0방향으로 평균을 구해 데이터프레임에 추가
print(average)
````
